```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
totalschools = school_data_complete['school_name'].nunique()
totalstudents = school_data_complete['Student ID'].nunique()
budget = school_data_complete['budget'].unique()
totalbudget = budget.sum()
avg_math = round(school_data_complete['math_score'].mean(),2)
avg_reading = round(school_data_complete['reading_score'].mean(),2)

math_70_df = school_data_complete.loc[school_data_complete['math_score']>=70]
math_70 = math_70_df['Student ID'].count()
math_pct = round(math_70 / totalstudents * 100,2)

reading_70_df = school_data_complete.loc[school_data_complete['reading_score']>=70]
reading_70 = reading_70_df['Student ID'].count()
reading_pct = round(reading_70 / totalstudents * 100,2)

math_read_70_df = school_data_complete.loc[(school_data_complete['reading_score']>=70) & (school_data_complete['math_score']>=70)]
math_read_70 = math_read_70_df['Student ID'].count()
math_read_pct = round(math_read_70 / totalstudents * 100,2)

district_sum_df = pd.DataFrame({'Total Schools': [totalschools], 
                               'Total Students': [totalstudents], 
                                'Total Budget': [totalbudget],
                               'Average Math Score': [avg_math],
                               'Average Reading Score': [avg_reading],
                                '% Passing Math': [math_pct], 
                                '% Passing Reading': [reading_pct],
                                '% Overall Passing': [math_read_pct] })
district_sum_df['Total Budget'] = district_sum_df['Total Budget'].apply(lambda x: "${:,.2f}".format(x))


district_sum_df
```

```python
grouped_school = school_data_complete.groupby(['school_name', 'type', 'size', 'budget'])

avg_math = grouped_school['math_score'].mean()
avg_reading = grouped_school['reading_score'].mean()

math_pass = grouped_school['math_score'].apply(lambda x: x[x >= 70].count())
reading_pass = grouped_school['reading_score'].apply(lambda x: x[x >= 70].count())
overall_pass = school_data_complete[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)].groupby(['school_name', 'type', 'size', 'budget'])['type'].count()

grouped_school_df = pd.DataFrame({'Ave Math': avg_math, 'Ave Reading': avg_reading, 'Number Math Pass': math_pass, 'Number Reading Pass': reading_pass, 'Overall Pass': overall_pass})
grouped_school_df
```

```python
grouped_school_df = grouped_school_df.reset_index()
grouped_school_df['% Math Pass'] = (grouped_school_df['Number Math Pass']/grouped_school_df['size'])*100
grouped_school_df['% Reading Pass'] = (grouped_school_df['Number Reading Pass']/grouped_school_df['size'])*100
grouped_school_df['% Overall Pass'] = (grouped_school_df['Overall Pass']/grouped_school_df['size'])*100
grouped_school_df['Per student budget'] = grouped_school_df['budget']/grouped_school_df['size']

grouped_school_df
```

```python
del grouped_school_df['Number Math Pass']
del grouped_school_df['Number Reading Pass']
del grouped_school_df['Overall Pass']
grouped_school_df.set_index('school_name')
```
